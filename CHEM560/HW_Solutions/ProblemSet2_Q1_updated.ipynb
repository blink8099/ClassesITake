We know that the x, y, z matrices can be expressed in the bra-ket notation:

X matrix:
$$  \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} =  \begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix} + \begin{bmatrix} 0 & 0 \\ 1 & 0 \end{bmatrix} =  \begin{bmatrix} 1 \\ 0 \end{bmatrix}  \begin{bmatrix} 0 & 1 \end{bmatrix} + \begin{bmatrix} 0 \\ 1 \end{bmatrix}  \begin{bmatrix} 1 & 0 \end{bmatrix} =  \vert 0 \rangle \langle 1 \vert + \vert 1 \rangle \langle 0 \vert  $$

Y matrix:
$$  \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} =  \begin{bmatrix} 0 & -i \\ 0 & 0 \end{bmatrix} + \begin{bmatrix} 0 & 0 \\ i & 0 \end{bmatrix} = -i \begin{bmatrix} 1 \\ 0 \end{bmatrix}  \begin{bmatrix} 0 & 1 \end{bmatrix} + i\begin{bmatrix} 0 \\ 1 \end{bmatrix}  \begin{bmatrix} 1 & 0 \end{bmatrix} =  -i \vert 0 \rangle \langle 1 \vert + i \vert 1 \rangle \langle 0 \vert  $$

Z matrix:
$$  \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} =  \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix} + \begin{bmatrix} 0 & 0 \\ 0 & -1 \end{bmatrix} =  \begin{bmatrix} 1 \\ 0 \end{bmatrix}  \begin{bmatrix} 1 & 0 \end{bmatrix} - \begin{bmatrix} 0 \\ 1 \end{bmatrix}  \begin{bmatrix} 0 & 1 \end{bmatrix} =  \vert 0 \rangle \langle 0 \vert - \vert 1 \rangle \langle 1 \vert  $$

Then we can use our observed probility to measure the expectation values for x, y, and z basis.

For X basis measurement, since qiskit only allow measureing in z basis, we first apply a H gate. We then observe: $\vert \psi \rangle =  \sqrt{1} \vert 0 \rangle $. Since the masurement is in the z-basis, we get:
$$ \langle X \rangle = \langle \psi \vert Z \vert \psi \rangle = \langle \psi \lvert (\vert 0 \rangle \langle 0 \vert - \vert 1 \rangle \langle 1 \vert )\rvert \psi \rangle = (\sqrt1)^2 = 1 $$ 


For Y basis measurement, since qiskit only allow measureing in z basis, we first apply a S+ gate then a H gate. We then observe: $\vert \psi \rangle =  \sqrt{0.502} \vert 0 \rangle + \sqrt{0.498} \vert 1 \rangle $. Since the masurement is in the z-basis, we get:
$$ \langle Y \rangle = \langle \psi \vert Z \vert \psi \rangle = \langle \psi \lvert (\vert 0 \rangle \langle 0 \vert - \vert 1 \rangle \langle 1 \vert )\rvert \psi \rangle = (\sqrt{0.502})^2 - (\sqrt{0.498})^2 = 0.004 $$ 

For z basis measurement, we observe: $\vert \psi \rangle =  \sqrt{0.508} \vert 0 \rangle + \sqrt{0.492} \vert 1 \rangle $. Since the masurement is in the z-basis, we get:
$$ \langle Z \rangle = \langle \psi \vert Z \vert \psi \rangle = \langle \psi \lvert (\vert 0 \rangle \langle 0 \vert - \vert 1 \rangle \langle 1 \vert )\rvert \psi \rangle = (\sqrt{0.508})^2 - (\sqrt{0.492})^2 = 0.016 $$ 

From the expectation values, density matrix $\rho$ can be calculated as:

$$ \rho = \dfrac{ \mathbf{I} + 1 \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} + 0.004 \begin{bmatrix} 0 & i \\ i & 0 \end{bmatrix} + 0.016 \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} }{2}  $$


```python
mat = np.matrix([[0.5+ 0.008, 0.5+0.002j], [0.5+0.002j, 0.5+ 0.008]])
rho = DensityMatrix(mat,dims=(2,))
```

```python
plot_bloch_multivector(rho)
```

```python
plot_state_city(rho)
```

```python
from numpy.linalg import matrix_power
from scipy.linalg import sqrtm
```

```python
pure_state = np.matrix([[0.5, 0.5], [0.5, 0.5]])
pure_state_oneHalf = sqrtm(pure_state)
```

```python
fid_1c_aer_simulator = np.trace(sqrtm((pure_state_oneHalf.dot(rho)).dot(pure_state_oneHalf)))
print(fid_1c_aer_simulator)
```

Note that the fidelity here is slightly larger than one. This is because of numerical noise. In reality this should never be larger than 1

We know that the x, y, z matrices can be expressed in the bra-ket notation:

X matrix:
$$  \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} =  \begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix} + \begin{bmatrix} 0 & 0 \\ 1 & 0 \end{bmatrix} =  \begin{bmatrix} 1 \\ 0 \end{bmatrix}  \begin{bmatrix} 0 & 1 \end{bmatrix} + \begin{bmatrix} 0 \\ 1 \end{bmatrix}  \begin{bmatrix} 1 & 0 \end{bmatrix} =  \vert 0 \rangle \langle 1 \vert + \vert 1 \rangle \langle 0 \vert  $$

Y matrix:
$$  \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} =  \begin{bmatrix} 0 & -i \\ 0 & 0 \end{bmatrix} + \begin{bmatrix} 0 & 0 \\ i & 0 \end{bmatrix} = -i \begin{bmatrix} 1 \\ 0 \end{bmatrix}  \begin{bmatrix} 0 & 1 \end{bmatrix} + i\begin{bmatrix} 0 \\ 1 \end{bmatrix}  \begin{bmatrix} 1 & 0 \end{bmatrix} =  -i \vert 0 \rangle \langle 1 \vert + i \vert 1 \rangle \langle 0 \vert  $$

Z matrix:
$$  \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} =  \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix} + \begin{bmatrix} 0 & 0 \\ 0 & -1 \end{bmatrix} =  \begin{bmatrix} 1 \\ 0 \end{bmatrix}  \begin{bmatrix} 1 & 0 \end{bmatrix} - \begin{bmatrix} 0 \\ 1 \end{bmatrix}  \begin{bmatrix} 0 & 1 \end{bmatrix} =  \vert 0 \rangle \langle 0 \vert - \vert 1 \rangle \langle 1 \vert  $$

Then we can use our observed probility to measure the expectation values for x, y, and z basis.

For X basis measurement, since qiskit only allow measureing in z basis, we first apply a H gate. We then observe: $\vert \psi \rangle =  \sqrt{0.894} \vert 0 \rangle + \sqrt{0.106} \vert 1 \rangle $. Since the masurement is in the z-basis, we get:
$$ \langle X \rangle = \langle \psi \vert Z \vert \psi \rangle = \langle \psi \lvert (\vert 0 \rangle \langle 0 \vert - \vert 1 \rangle \langle 1 \vert )\rvert \psi \rangle = (\sqrt1)^2 = 0.788 $$ 


For Y basis measurement, since qiskit only allow measureing in z basis, we first apply a S+ gate then a H gate. We then observe: $\vert \psi \rangle =  \sqrt{0.475} \vert 0 \rangle + \sqrt{0.525} \vert 1 \rangle $. Since the masurement is in the z-basis, we get:
$$ \langle Y \rangle = \langle \psi \vert Z \vert \psi \rangle = \langle \psi \lvert (\vert 0 \rangle \langle 0 \vert - \vert 1 \rangle \langle 1 \vert )\rvert \psi \rangle = (\sqrt{0.475})^2 - (\sqrt{0.525})^2 = −0.05 $$ 

For z basis measurement, we observe: $\vert \psi \rangle =  \sqrt{0.481} \vert 0 \rangle + \sqrt{0.519} \vert 1 \rangle $. Since the masurement is in the z-basis, we get:
$$ \langle Z \rangle = \langle \psi \vert Z \vert \psi \rangle = \langle \psi \lvert (\vert 0 \rangle \langle 0 \vert - \vert 1 \rangle \langle 1 \vert )\rvert \psi \rangle = (\sqrt{0.481})^2 - (\sqrt{0.519})^2 = −0.038 $$ 

From the expectation values, density matrix $\rho$ can be calculated as:

$$ \rho = \dfrac{ \mathbf{I} + 0.788 \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} - 0.05 \begin{bmatrix} 0 & i \\ i & 0 \end{bmatrix} - 0.038 \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} }{2}  $$
